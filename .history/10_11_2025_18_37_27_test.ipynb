## Data Description

`players.csv` has one row per player and includes the main variables: **experience** (Amateur, Veteran, or Pro), **subscribe** (TRUE/FALSE), **hashedEmail** (unique player ID), **played_hours** (total hours spent playing), **name**, **gender**, and **Age** (in years).  

`sessions.csv` lists individual play sessions for each player. However, for this analysis, I focus on `players.csv` because it is easier to summarize and directly relates to my question about which players subscribe to the newsletter. I calculated one-number summaries (means) for the numeric variables in `players.csv`, created a histogram of played hours, and plotted the subscription rate by experience level. While calculating the means, some values for **Age** were missing (shown as `NA`). This was handled using the argument `na.rm = TRUE`, which removes missing values so that the average can still be computed correctly.


```r
players |>
  select(played_hours, Age) |>
  summarise(across(c(played_hours, Age), ~ round(mean(.x, na.rm = TRUE), 2))) |>
  pivot_longer(cols = c(played_hours, Age),
               names_to = "variable",
               values_to = "mean_2dp") |>
  arrange(variable)
```
